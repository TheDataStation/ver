```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200
```

```python
times = []
with open("times.log", "r") as f:
    lines = f.readlines()
    for line in lines:
        times.append(float(line))
sample_portions = ["0.25", "0.5", "0.75", "1.0"]

plt.bar(sample_portions, times, width = 0.4)
plt.xlabel("sample portion")
plt.ylabel("run time(s)")
```

```python
num_views = []
compatible = []
contained = []
complementary = []
contradictory = []

with open("results.log", "r") as f:
    lines = f.readlines()
    for line in lines:
        nums = line.strip().split()
        num_views.append(int(nums[0]))
        compatible.append(int(nums[1]))
        contained.append(int(nums[2]))
        complementary.append(int(nums[3]))
        contradictory.append(int(nums[4]))

x = np.arange(len(sample_portions))  # the label locations
width = 0.15  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2*width, num_views, width, label='num views')
rects2 = ax.bar(x - width, compatible, width, label='num compatible group')
rects3 = ax.bar(x, contained, width, label='num largest contained view')
rects4 = ax.bar(
    x + width, contradictory, width, label='num contradictions')
rects5 = ax.bar(x + 2*width, complementary, width, label='num complementary pairs')

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel('run time(s)')
ax.set_xlabel('sample portion')
ax.set_xticks(x, sample_portions)
ax.legend()

ax.bar_label(rects1, padding=2)
ax.bar_label(rects2, padding=2)
ax.bar_label(rects3, padding=2)
ax.bar_label(rects4, padding=2)
ax.bar_label(rects5, padding=2)

# ax.set_yscale('log')

fig.tight_layout()

plt.show()
```
